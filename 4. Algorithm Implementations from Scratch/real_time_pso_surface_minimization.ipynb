```python
import numpy as np
from matplotlib import pyplot as plt
import random
import pandas as pd
import time
%matplotlib qt
```

```python
def func(x,y):
    z=np.arctan(x**2+y**2)+np.sin(x+y)
    return z
#
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)
x, y = np.meshgrid(x, y)
z = np.arctan(x**2+y**2)+np.sin(x+y)

# 3D yüzey grafiğini çizme
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis', alpha=0.7)
plt.tight_layout()
cset = ax.contourf(x, y, z, zdir='z', offset=np.min(z), cmap='viridis', alpha=0.5)

#

particle_quantity=50
particle_x=np.random.uniform(-8, 8, particle_quantity)
particle_y=np.random.uniform(-8, 8, particle_quantity)
particle_z=func(particle_x,particle_y)

best_particle_id=np.argmin(particle_z)
best_particle=(particle_x[best_particle_id] , particle_y[best_particle_id], particle_z[best_particle_id])

particles_best_attempts=(particle_x,particle_y,particle_z)
#
speed_x=np.random.uniform(-0.2, 0.2, particle_quantity)
speed_y=np.random.uniform(-0.2, 0.2, particle_quantity)
#
scatter = ax.scatter([particle_x],[particle_y],[particle_z], color='red', s=50, label='Noktalar')

w,c1,c2,r1,r2,r11,r12=0.4,1.5,1.5,random.uniform(0, 1),random.uniform(0, 1),random.uniform(0, 1),random.uniform(0, 1)

while True:
    for a in range(0,particle_x.shape[0]):
        if particles_best_attempts[2][a]>particle_z[a]:
            particles_best_attempts[0][a]=particle_x[a]
            particles_best_attempts[1][a]=particle_y[a]
            particles_best_attempts[2][a]=particle_z[a]
            #
        speed_x[a]=(w*speed_x[a])+(c1*r1*(particles_best_attempts[0][a]-particle_x[a]))+(c2*r2*(particle_x[best_particle_id]-particle_x[a]))
        speed_y[a]=(w*speed_y[a])+(c1*r11*(particles_best_attempts[1][a]-particle_y[a]))+(c2*r12*(particle_y[best_particle_id]-particle_y[a]))
        
        r1,r2,r11,r12=random.uniform(0, 1),random.uniform(0, 1),random.uniform(0, 1),random.uniform(0, 1)

    particle_x+=speed_x
    particle_y+=speed_y
    #
    particle_z=func(particle_x,particle_y)
    best_particle_id=np.argmin(particle_z)
    best_particle=(particle_x[best_particle_id] , particle_y[best_particle_id], particle_z[best_particle_id])
    
    scatter.remove()
    scatter = ax.scatter([particle_x],[particle_y],[particle_z], color='red', s=50)
    
    if np.fabs(speed_x.max())<0.01 and np.fabs(speed_y.max())<0.01:
        break
    plt.pause(0.9)


    

print(particle_x[-1],particle_y[-1])
```
