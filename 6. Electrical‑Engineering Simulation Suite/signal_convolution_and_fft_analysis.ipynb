```python
import numpy as np
import matplotlib.pyplot as plt






time = np.arange(0, 20, 0.001)
signal_sine = np.sin(time)
signal_exp = np.exp(-time)

convolution_result = np.convolve(signal_sine, signal_exp, mode='same')


plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(time, signal_sine, label='Sine Dalga (f=10 Hz)')
plt.legend()
plt.title('Sine Dalga')
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(time, signal_exp, label='exp(-t)')
plt.legend()
plt.title('exp(-t)')
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(time, convolution_result, label='KonvolÃ¼syon Sonucu')
plt.legend()
plt.title('KonvolÃ¼syon Sonucu')
plt.grid()

plt.tight_layout()
plt.show()
```

```python


ps=1/500

t = np.arange(0, 2, ps)

signal1 = 3 * np.cos(2 * np.pi * 30 * t + 0.6)
signal2 = 3 * np.cos(2 * np.pi * 45 * t + -0.8)
signal3 = np.cos(2 * np.pi * 70 * t + 2)

sumsignals = signal1 + signal2 + signal3

N = len(t)
fs = 1 / (t[1] - t[0])
S = np.fft.fft(sumsignals)
S_oneSide = S[:N // 2]
f = fs * np.arange(N // 2) /N
S_mag = np.abs(S_oneSide) / (N / 2)

plt.figure(figsize=(16, 6))

plt.subplot(3, 1, 1)
plt.plot(t, sumsignals)
plt.title('Time Domain Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(f, S_mag)
plt.title('Frequency Domain Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(f, np.angle(S_oneSide))
plt.title('Phase Angles')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase Angle')

plt.tight_layout()
plt.show()
```
