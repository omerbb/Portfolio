```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 5, 0.01)
x = 10 * np.exp(-t) - 5 * np.exp(0.5 * t)
plt.figure(figsize=(20, 8))
plt.plot(t, x)
plt.title('Continuous Time Signal: $x(t) = 10e^{-t} - 5e^{0.5t}$')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

```python
def damped_sinusoidal_signal(t, a):
    x = 20 * np.sin(2 * np.pi * 1000 * t - np.pi/3) * np.exp(a * t)
    return x


t = np.linspace(-2e-3, 2e-3, 1000)

a = [250, 500, 750, 1000]

plt.figure(figsize=(20, 8))


for b in a:
    x_t = damped_sinusoidal_signal(t, b)
    plt.plot(t * 1e3, x_t, label=f'a = {b}')
plt.xlabel('Time (ms)')
plt.ylabel('x(t)')
plt.title('Exponentially Damped Sinusoidal Signal for Different a Values')
plt.legend()
plt.grid(True)
plt.show()


```

```python
def delta_continuous(t, t0):
    delta = np.zeros_like(t)
    delta[t == t0] = 1
    return delta

def delta_discrete(n, n0):
    delta = np.zeros_like(n)
    delta[n == n0] = 1
    return delta


t = np.arange(-5, 10, 1)
n = np.arange(-5, 10, 1)



y1 = delta_continuous(t, 2)
y2 = delta_discrete(n, 5)


plt.figure(figsize=(20, 8))

plt.subplot(2, 1, 1)
plt.stem(t, y1,"-b",markerfmt="|b",basefmt="b")
plt.title('Delta function for continuous signal')
plt.xlabel('t')
plt.ylabel('δ(t - 2)')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(n, y2,"o")
plt.title('Delta function for discrete signal')
plt.xlabel('n')
plt.ylabel('δ[n - 5]')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python

def x_t(t):
    if -1 <= t <= 0:
        return -t + 1
    elif 0 <= t <= 1:
        return t + 1
    else:
        return 0


t = np.arange(-8, 8, 0.001)



y1 = [x_t(t - 5) for t in t]
y2 = [x_t(t + 5) for t in t]
y3 = [x_t(2 * t - 4) for t in t]
y4 = [-2 * x_t(-2 * t + 5) for t in t]


plt.figure(figsize=(20, 8))
plt.subplot(2, 2, 1)
plt.plot(t, y1)
plt.title('x(t-5)')
plt.xlabel('t')
plt.ylabel('x(t-5)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(t, y2)
plt.title('x(t+5)')
plt.xlabel('t')
plt.ylabel('x(t+5)')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(t, y3)
plt.title('x(2t-4)')
plt.xlabel('t')
plt.ylabel('x(2t-4)')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(t, y4)
plt.title('-2x(-2t+5)')
plt.xlabel('t')
plt.ylabel('-2x(-2t+5)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
