```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.signal import filtfilt
```

Soru 2

```python
def dtsignal(n,f):
    if -1/(2*f)<= n<=1/(2*f):
        return np.cos(2*np.pi*f*n)
    else:
        return 0

F=0.1
y=[]


x = np.arange(-1/(2*F), 1/(2*F), 0.1)
for a in x:
    y.append(dtsignal(a,F))

plt.figure(figsize=(15, 5))
plt.grid()
plt.plot(x,y,".")
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
```

```python

t = np.linspace(0, 1, 1000)
noisy_signal = 2 * np.sin(2 * np.pi * 1500 * t) + 0.5 * np.sin(2 * np.pi * 3000 * t) + 0.5 * np.random.randn(len(t))


filter_order = 4
cutof_frequency = 1000
sampling_frequency = 5000

b, a = signal.butter(filter_order, cutof_frequency / (sampling_frequency / 2), 'low')

filtered_signal = filtfilt(b, a, noisy_signal)

w, h = signal.freqz(b, a, worN=8000)
frequencies = (sampling_frequency / 2) * (w / np.pi)


plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.plot(t,noisy_signal)
plt.title('(noisy signal)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.subplot(2, 2, 2)
plt.plot(t,filtered_signal)
plt.title('(filtered signal)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.subplot(2, 2, (3,4))
plt.plot(frequencies, 20 * np.log10(abs(h)))
plt.title('(frequency response)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.tight_layout()
plt.show()
```
