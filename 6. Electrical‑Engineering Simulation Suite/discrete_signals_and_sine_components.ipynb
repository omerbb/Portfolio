```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(-20,20)
y=[]
for a in x:
    if 1<=a<=10:
        y.append(10)
    else:
        y.append(0)
plt.figure(figsize=(15, 5))
plt.plot(x,y)
plt.title('Pulse)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
```

```python
sampling_frequency = 1000  
amplitude = 10
t = np.arange(0, 1, 1 / sampling_frequency)

sin_wave1 = amplitude * np.sin(2 * np.pi * 10 * t)
sin_wave2 = amplitude * np.sin(2 * np.pi * 30 * t)
sin_wave3 = amplitude * np.sin(2 * np.pi * 70 * t)

sum_wave = sin_wave1 + sin_wave2 + sin_wave3

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(t, sin_wave1)
plt.title('(10 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 2)
plt.plot(t, sin_wave2)
plt.title('(30 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 3)
plt.plot(t, sin_wave3)
plt.title('(70 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 4)
plt.plot(t, sum_wave)
plt.title('Sum of Sin Waves')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')


plt.tight_layout()
plt.show()
```

```python
def u(n):
    if n>=0:
        return 1
    else:
        return 0
    
def ddelt(n):
    if n ==0:
        return 1
    else:
        return 0
    
x = range(-10,10)
y1=[]
y2=[]

for n in x:
    y1.append(u(n)-2*(u(n-1))+u(n-4))
    y2.append(ddelt(n+1)-ddelt(n)+u(n+1)-u(n-2))
    
plt.figure(figsize=(15, 5))
plt.subplot(2, 2, 1)
plt.plot(x, y1, "*")
plt.title('u[n] âˆ’ 2u[n âˆ’ 1] + u[n âˆ’ 4]')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 2, 2)
plt.plot(x, y2, "*")
plt.title('Î´[n + 1] âˆ’ Î´[n] + u[n + 1] âˆ’ u[n âˆ’ 2]')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
