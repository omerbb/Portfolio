```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write, read

Fs = 11025
t = np.arange(-0.1, 0.1, 1/Fs)
f_low = 697
f_high = 1209
x_t_1 = np.sin(2 * np.pi * f_low * t) + np.sin(2 * np.pi * f_high * t)


write('dtmf1.wav', Fs, np.int16(x_t_1 * 32767))
y, _ = read('dtmf1.wav')

plt.figure(figsize=(12, 6))
plt.plot(t, x_t_1)
plt.title('Button 1 Sinusoidal Dalga Formu')
plt.xlabel('Zaman (s)')
plt.ylabel('AmplitÃ¼d')
plt.show()
```

```python
step_size = 0.001



time = np.arange(0, 2, step_size)
signal_sine = np.sin(2 * np.pi * 10 * time)
signal_cosine = np.cos(2 * np.pi * 15 * time)

convolution_result = np.convolve(signal_sine, signal_cosine, mode='same') * step_size


plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(time, signal_sine, label='Sine Dalga (f=10 Hz)')
plt.legend()
plt.title('Sine Dalga')

plt.subplot(3, 1, 2)
plt.plot(time, signal_cosine, label='Cosine Dalga (f=15 Hz)')
plt.legend()
plt.title('Cosine Dalga')

plt.subplot(3, 1, 3)
plt.plot(time, convolution_result, label='KonvolÃ¼syon Sonucu')
plt.legend()
plt.title('KonvolÃ¼syon Sonucu')

plt.tight_layout()
plt.show()
```
